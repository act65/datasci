```python
import os
import csv
import collections
```

```python
def get_data_paths(basedir):
    datafiles = []
    for root, dirs, files in os.walk(basedir):
        if len(files) <= 3:
            for f in files:
                if f.endswith('.csv'):
                    datafiles.append(os.path.join(root, f))
    return datafiles

def num(s):  # what about bools? or categoricals?
    try:
        return int(s)
    except ValueError:
        try:
            return float(s)
        except ValueError:
            return s
```

```python
basedir = '/home/act65/.kaggle/datasets/uciml/'

fnames = get_data_paths(basedir)
print(len(fnames))
```

```python

ds = collections.namedtuple('Dataset', ['Title', 'Column_ids', 'n_rows'])
```

```python
datasets = []
for fname in fnames:
    with open(fname, encoding="utf-8") as f:
        try:
            reader = csv.reader(f)
            column_names = next(reader)
            first_row = next(reader)
#             print(first_row)
#             break
            datasets.append(ds(fname, 
                               list(zip(column_names, [type(num(r)) for r in first_row])), 
                               len([r[0] for r in reader]) + 1,
                               ))
            # also want their dtype?
        except UnicodeDecodeError:  # not all files are in the same format?
            pass
```
